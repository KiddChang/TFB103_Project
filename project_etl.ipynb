```python
db = client.get_database('raw_data_for_project')
```

```python
collection = db.ifoodie
```

```python
result=collection.find({})
origin_ifoodie=pd.json_normalize([json for json in result])
```

```python
ifoodie= origin_ifoodie.copy()
ifoodie
```

```python
#pd.set_option('display.max_rows',None)
```

```python
ifoodiedis=ifoodie.drop_duplicates(subset='articleUrl',keep='first')
```

```python
ifoodiedis['Ave_consumption']=ifoodiedis.loc[ifoodiedis['Ave_consumption'].str.contains('均消'), 'Ave_consumption']
```

```python
ifoodiedis['Ave_consumption']=ifoodiedis['Ave_consumption'].str.split(' ',expand=True)[1].str.split('$',expand=True)[1]
```

```python
ifoodiedis['Commentcount']=ifoodiedis['Commentcount'].str.split('則評論',expand=True)[0]
```

```python
ifoodiedis['Com_comment']=ifoodiedis['Com_comment'].apply(lambda x: re.findall('[\u4e00-\u9fa5]+\d*[\u4e00-\u9fa5]+', "".join(x)))
```

```python
ifoodiedis.rename(columns={'_id':'id',"Shop_name":"餐廳名稱","address":"地址","Point":"評分","Commentcount":"評論數","Ave_consumption":"均消","articleUrl":"文章連結","Com_comment":"內文","image":"圖片連結"},inplace=True)
```

```python
ifoodiedis
```

```python
ifoodieFinal = pd.DataFrame(ifoodiedis, columns=['id','餐廳名稱','地址','內文','評分','均消','文章連結','圖片連結','評論數'])
```
